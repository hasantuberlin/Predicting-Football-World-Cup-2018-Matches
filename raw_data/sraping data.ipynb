```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_countries_by_government_budget').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
links = My_table.findAll('td')
linkss=My_table.findAll('a')
linkss
```

```python
Countries = []
for link in linkss:
    if link.get('title')==None:
        continue
    Countries.append(link.get('title'))
    
print(Countries)
```

```python
links
links[0].a.text
```

```python
import pandas as pd
df = pd.DataFrame()
df['Country'] = Countries

df
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import sys
reload(sys)
sys.setdefaultencoding('utf8')
download_dir = "budget_final_v7.csv" #where you want the file to be downloaded to 

csv = open(download_dir, "w")
#"w" indicates that you're writing strings to the file
#columnTitleRow = "name, email\n"
columnTitleRow = "Rank, Country, Revenues, Expenditures, Surplus\n"
csv.write(columnTitleRow)
URL = "https://en.wikipedia.org/wiki/List_of_countries_by_government_budget"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
for items in soup.find('table', class_='wikitable').find_all('tr')[1:]:
    data = items.find_all(['th','td'])
    try:
        Rank = data[0].text
        Country = data[1].a.text
        Revenues = data[2].text
        Expenditures = data[3].text
        Surplus = data[4].text
        #row = name + "," + email + "\n"
        row = Rank + ", " + Country + ", "+ Revenues +", "+ Expenditures +", "+ Surplus+"\n"
    except IndexError:pass
    csv.write(row)
    #print("{}|{}|{}".format(Rank,Country,Revenues))
    

   
```
